```python
# from googletrans import Translator
def translate_text(text):
    translator = Translator()
    translated = translator.translate(text, src='fa', dest='ar')
    return translated # translation.text

translation = translate_text(input_text)
print(translation)
```

```python

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
ARTICLE =translation.text

print(summarizer(ARTICLE, max_length=130, min_length=30, do_sample=False))
keywords_Extracter=summarizer(ARTICLE, max_length=130, min_length=30, do_sample=False)
```

```python
# بارگذاری مدل عمومی
model_name = "csebuetnlp/mT5_multilingual_XLSum"
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# متن عربی که نیاز به خلاصه‌سازی دارد
arabic_text = translation.text


# خلاصه‌سازی متن
def summarize_arabic_text(text, max_length=100, min_length=55):
    inputs = tokenizer(text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(
        inputs["input_ids"],
        max_length=max_length,
        min_length=min_length,
        length_penalty=2.0,
        num_beams=4,
        early_stopping=True
    )
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# اجرای خلاصه‌سازی
arabic_summary = summarize_arabic_text(arabic_text)
print("خلاصه متن عربی:\n", arabic_summary)
```

```python
# from googletrans import Translator
def translate_text(text):
    translator = Translator()
    translated = translator.translate(text, src='ar', dest='fa')
    return translated # translation.text

translation = translate_text(arabic_summary)
print(translation)
```
